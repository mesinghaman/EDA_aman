```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bank_marketing_test.csv')

# Convert month and day_of_week to categorical types for ordering
data['month'] = pd.Categorical(data['month'], categories=['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'], ordered=True)
data['day_of_week'] = pd.Categorical(data['day_of_week'], categories=['mon', 'tue', 'wed', 'thu', 'fri'], ordered=True)

# Monthly Contact Volume
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='month', palette='Blues')
plt.title("Number of Contacts by Month")
plt.xlabel("Month")
plt.ylabel("Number of Contacts")
plt.show()

# Weekly Contact Volume
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='day_of_week', palette='Purples')
plt.title("Number of Contacts by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Contacts")
plt.show()

# Monthly Contact Duration Trend
plt.figure(figsize=(10, 5))
sns.boxplot(data=data, x='month', y='duration', palette='coolwarm')
plt.title("Distribution of Duration by Month")
plt.xlabel("Month")
plt.ylabel("Duration (seconds)")
plt.show()

# Outcome by Month (Subscription outcome 'y')
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='month', hue='y', palette='Set2')
plt.title("Distribution of Subscription Outcome (y) by Month")
plt.xlabel("Month")
plt.ylabel("Count")
plt.legend(title="Subscription Outcome")
plt.show()
```

 perform the EDA for 1 Dimension analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bank_marketing_test.csv')

# Summary statistics for numerical features
print("Summary Statistics for Numerical Features:")
print(data.describe())

# Numerical features for analysis
numerical_features = ["age", "duration", "campaign", "pdays", "previous", "emp.var.rate",
                      "cons.price.idx", "cons.conf.idx", "euribor3m", "nr.employed"]

# Histograms and Boxplots for numerical features
for feature in numerical_features:
    plt.figure(figsize=(14, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data[feature], kde=True, color='skyblue')
    plt.title(f"Histogram of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[feature], color='lightcoral')
    plt.title(f"Boxplot of {feature}")
    plt.xlabel(feature)

    plt.show()

# Frequency counts for categorical features
categorical_features = ["job", "marital", "education", "default", "housing", "loan", "contact",
                        "month", "day_of_week", "poutcome", "y"]

# Count plots for categorical features
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=data, x=feature, palette='viridis')
    plt.title(f"Count Plot of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

# Display unique value counts for categorical features
print("Unique Value Counts for Categorical Features:")
for feature in categorical_features:
    print(f"\n{feature}:\n{data[feature].value_counts()}")
```

perform the EDA for 2D and N-D analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bank_marketing_test.csv')

print(data.info())
print(data.describe(include='all'))
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])

numerical_features = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate',
                      'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
data[numerical_features].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features')
plt.show()

categorical_features = ['job', 'marital', 'education', 'default', 'housing',
                        'loan', 'contact', 'month', 'day_of_week', 'poutcome', 'y']
plt.figure(figsize=(15, 10))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 4, i)
    sns.countplot(y=data[feature], order=data[feature].value_counts().index)
    plt.title(f'{feature} Distribution')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(x='y', y=feature, data=data)
    plt.title(f'{feature} vs y')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 4, i)
    sns.countplot(x='y', hue=feature, data=data)
    plt.title(f'{feature} vs y')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
correlation_matrix = data[numerical_features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Features')
plt.show()

selected_features = ['age', 'duration', 'campaign', 'previous', 'euribor3m', 'nr.employed', 'y']
sns.pairplot(data[selected_features], hue='y', diag_kind='kde')
plt.suptitle('Pair Plot of Selected Features')
plt.show()
```
