```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bank_marketing_test.csv')
X = dataset.iloc[:, [0, 10]].values
```

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    #as Kmean is a stochastic process, there is always a chance to stuck in local minima
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training

```
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

## Visualisization

```
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('age')
plt.ylabel('duration')
plt.legend()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```
df = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/bank_marketing_test.csv')
```

```
plt.figure(figsize=(8,5))
plt.title("Time Duration distribution",fontsize=16)
plt.xlabel ("duration",fontsize=14)
plt.grid(True)
plt.hist(df['duration'],color='blue',edgecolor='k')
plt.show()
```

```
plt.figure(figsize=(8,5))
plt.title("Age distribution",fontsize=16)
plt.xlabel ("age",fontsize=14)
plt.grid(True)
plt.hist(df['age'],color='Red',edgecolor='k')
plt.show()
```

```

#Dendograms
X = df.iloc[:,[0,10]].values
```

```
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(15,6))
plt.title('Dendrogram')
plt.xlabel('Age')
plt.ylabel('Euclidean distances')
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.show()
```
