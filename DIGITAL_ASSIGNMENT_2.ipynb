```python
columns_to_drop = ['fuel-system', 'bore']

columns_to_drop = [col for col in columns_to_drop if col in data.columns] or ['loan', 'contact']

data = data.drop(columns=columns_to_drop)

print(data)
```

```python
data.info()
```

```python
missing_values = data.isna().sum()

print(missing_values)
```

```python
data['age'] = data['age'].fillna(data['age'].mean())

print(data)
```

```python
mean_age = data['age'].mean()
median_age = data['age'].median()
mode_age = data['age'].mode()[0]

print(f"Mean Height: {mean_age}")
print(f"Median Height: {median_age}")
print(f"Mode Height: {mode_age}")
```

```python
range_age = data['age'].max() - data['age'].min()
variance_age = data['age'].var()
std_dev_age = data['age'].std() # remove sage and add std() to get the standard deviation.
print(f"Range of Height: {range_age}")
print(f"Variance of Height: {variance_age}")
print(f"Standard Deviation of Height: {std_dev_age}")
```

```python
Q1 = data['age'].quantile(0.25)
Q2 = data['age'].quantile(0.50)
Q3 = data['age'].quantile(0.75)


IQR = Q3 - Q1

print(f"Q1 (25th percentile): {Q1}")
print(f"Q2 (Median/50th percentile): {Q2}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
```

```python
descriptive_stats = data.describe()

print(descriptive_stats)
```

```python
from sklearn.preprocessing import LabelEncoder

le_marital = LabelEncoder()
le_job = LabelEncoder()

data['marital_encoded'] = le_marital.fit_transform(data['marital'])
data['job_encoded'] = le_job.fit_transform(data['job'])

# Calculate the correlation between the encoded columns
correlation = data['marital_encoded'].corr(data['job_encoded'])

# Print the correlation result
print(f"Correlation between marital and job: {correlation}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Create example data
np.random.seed(0)
data = pd.DataFrame({'variable_name': np.random.randn(100)})

# 1. Histogram
plt.figure(figsize=(8, 5))
plt.hist(data['variable_name'], bins=10, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Variable')
plt.ylabel('Frequency')
plt.show()

# 2. Density Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(data['variable_name'], fill=True, color='red', alpha=0.5)
plt.title('Density Plot')
plt.xlabel('Variable')
plt.ylabel('Density')
plt.show()

# 3. Histogram and Density
plt.figure(figsize=(8, 5))
sns.histplot(data['variable_name'], bins=10, kde=True, color='blue', edgecolor='black')
plt.title('Histogram and Density Plot')
plt.xlabel('Variable')
plt.ylabel('Density')
plt.show()

# 4. Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(y=data['variable_name'], color='green')
plt.title('Boxplot')
plt.ylabel('Variable')
plt.show()

# 5. Vertical Bar Plot
# Example data for vertical bar plot
bar_data = data['variable_name'].value_counts()
plt.figure(figsize=(8, 5))
bar_data.plot(kind='bar', color='purple')
plt.title('Vertical Bar Plot')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 6. Horizontal Bar Plot
plt.figure(figsize=(8, 5))
bar_data.plot(kind='barh', color='orange')
plt.title('Horizontal Bar Plot')
plt.xlabel('Frequency')
plt.ylabel('Value')
plt.show()

# 7. Pie Plot
plt.figure(figsize=(8, 5))
bar_data.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Reds')) # Change 'red' to 'Reds' which is a valid palette name
plt.title('Pie Plot')
plt.ylabel('')
plt.show()



# 9. Dot Plot
plt.figure(figsize=(8, 5))
sns.stripplot(x='variable_name', data=data, color='black', alpha=0.5)
plt.title('Dot Plot')
plt.xlabel('Variable')
plt.show()
```

```python
data = pd.read_csv('/content/bank_marketing_test.csv')
```

```python
mean_age = data['age'].mean()
median_age = data['age'].median()
mode_age = data['age'].mode()[0]

print(f'Mean Age: {mean_age}, Median Age: {median_age}, Mode Age: {mode_age}')
```

```python
variance_age = data['age'].var()
std_dev_age = data['age'].std()
range_age = data['age'].max() - data['age'].min()

print(f'Variance Age: {variance_age}, Standard Deviation Age: {std_dev_age}, Range Age: {range_age}')
```

```python
Q1 = data['age'].quantile(0.25)
Q3 = data['age'].quantile(0.75)
IQR = Q3 - Q1

print(f'Q1: {Q1}, Q3: {Q3}, IQR: {IQR}')
```

```python
contingency_table = pd.crosstab(data['job'], data['marital'])
print(contingency_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Stacked Bar Plot for 'job' and 'marital'
contingency_table.plot(kind='bar', stacked=True, color=['lightblue', 'lightgreen'])
plt.title('Stacked Bar Plot: Job vs. Marital Status')
plt.ylabel('Count')
plt.show()

# Grouped Bar Plot
sns.countplot(data=data, x='job', hue='marital')
plt.title('Grouped Bar Plot: Job vs. Marital Status')
plt.ylabel('Count')
plt.show()

# Scatter Plot
plt.scatter(data['age'], data['campaign'])
plt.title('Scatter Plot: Age vs. Campaign')
plt.xlabel('Age')
plt.ylabel('Campaign')
plt.show()

# Box Plot for `age` and `marital`
sns.boxplot(x='marital', y='age', data=data)
plt.title('Box Plot: Age by Marital Status')
plt.show()

# Violin Plot for `age` and `marital`
sns.violinplot(x='marital', y='age', data=data)
plt.title('Violin Plot: Age by Marital Status')
plt.show()

# Heatmap of correlation (only numerical values)
sns.heatmap(data.corr(numeric_only=True), cmap='coolwarm') # Added numeric_only=True to handle non-numerical values
plt.title('Heatmap of Correlation')
plt.show()

# Density Plot for age
sns.kdeplot(data['age'], fill=True, color='purple')
plt.title('Density Plot: Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

# Pairs Plot
sns.pairplot(data)
plt.title('Pairs Plot')
plt.show()
```
