**Discretization and binning**

```python
#To convert the preceding dataset into intervals, we can use the cut() method from pandas
data = {
    'age': [56, 41, 25, 35, 46, 55, 56, 40, 45, 53],
    'job': ['services', 'blue-collar', 'services', 'blue-collar', 'blue-collar', 'unknown', 'technician', 'blue-collar', 'blue-collar', 'admin.'],
    'marital': ['married', 'married', 'single', 'married', 'married', 'married', 'married', 'married', 'married', 'single'],
    'education': ['high.school', 'unknown', 'high.school', 'basic.6y', 'basic.6y', 'basic.4y', 'basic.4y', 'basic.9y', 'basic.9y', 'professional.course'],
    'default': ['no', 'no', 'no', 'no', 'unknown', 'unknown', 'no', 'no', 'no', 'no'],
    'housing': ['no', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'loan': ['yes', 'no', 'no', 'no', 'yes', 'no', 'no', 'yes', 'no', 'no'],
    'contact': ['telephone'] * 10,
    'month': ['may'] * 10,
    'day_of_week': ['mon'] * 10,
    'campaign': [1] * 10,
    'pdays': [999] * 10,
    'previous': [0] * 10,
    'poutcome': ['nonexistent'] * 10,
    'emp.var.rate': [1.1] * 10,
    'cons.price.idx': [93.994] * 10,
    'cons.conf.idx': [-36.4] * 10,
    'euribor3m': [4.857] * 10,
    'nr.employed': [5191.0] * 10,
    'y': ['no'] * 10
}

# Create DataFrame
dfx = pd.DataFrame(data)

# Define the bins and labels
bins = [0, 30, 40, 50, 60, 100]
labels = ['0-30', '31-40', '41-50', '51-60', '61+']

# Create a new column 'age_group' by cutting the 'age' column
dfx['age_group'] = pd.cut(dfx['age'], bins=bins, labels=labels, right=False)

# Display the DataFrame with the new age_group column
print(dfx)
```

```python
#We can check the number of values in each bin by using the pd.value_counts() method
# Create DataFrame
dfx = pd.DataFrame(data)

# Define the bins and labels
bins = [0, 30, 40, 50, 60, 100]
labels = ['0-30', '31-40', '41-50', '51-60', '61+']

# Create a new column 'age_group' by cutting the 'age' column
dfx['age_group'] = pd.cut(dfx['age'], bins=bins, labels=labels, right=False)

# Check the number of values in each bin
value_counts = pd.value_counts(dfx['age_group'])
print(value_counts)
```

```python
#We can also indicate the bin names by passing a list of labels
# Create DataFrame
dfx = pd.DataFrame(data)

# Define the bins and custom labels
bins = [0, 30, 40, 50, 60, 100]
bin_names = ['Youth', 'Young Adult', 'Middle Age', 'Senior', 'Elderly']

# Create a new column 'age_group' by cutting the 'age' column with custom labels
dfx['age_group'] = pd.cut(dfx['age'], bins=bins, labels=bin_names, right=False)

# Check the number of values in each bin
value_counts = pd.value_counts(dfx['age_group'])
print(value_counts)
```

```python
#it is essential to note that if we pass just an integer for our bins, it will compute equal-length bins based on the minimum and maximum values in the data.
# Create DataFrame
dfx = pd.DataFrame(data)

# Generate 40 random numbers between 0 and 1
random_scores = np.random.rand(40)

# Add the random scores to the DataFrame
dfx['random_score'] = np.random.choice(random_scores, size=len(dfx))

# Create equal-length bins (5 bins in this case)
bins = pd.cut(dfx['random_score'], 5, precision=2)

# Check the count of values in each bin
value_counts = pd.value_counts(bins)
print(value_counts)
```

```python
  #Pandas provides a qcut method that forms the bins based on sample quantiles
dfx = pd.DataFrame(data)

# Generate 2000 random numbers between 0 and 1
randomNumbers = np.random.rand(2000)

# Add the random numbers to the DataFrame as a new column
dfx['random_score'] = np.random.choice(randomNumbers, size=len(dfx))

# Use qcut to create quartiles
category3 = pd.qcut(dfx['random_score'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Add the quartile categories to the DataFrame
dfx['quartile'] = category3

# Display the DataFrame with the new quartile column
print(dfx[['random_score', 'quartile']])
```

```python
#Based on the number of bins, which we set to 4, it converted our data into four different categories.
dfx = pd.DataFrame(data)

# Generate 2000 random numbers between 0 and 1
randomNumbers = np.random.rand(2000)

# Add the random numbers to the DataFrame as a new column
dfx['random_score'] = np.random.choice(randomNumbers, size=len(dfx))

# Use qcut to create quartiles
category3 = pd.qcut(dfx['random_score'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Add the quartile categories to the DataFrame
dfx['quartile'] = category3

# Count the number of values in each category
category_counts = pd.value_counts(category3)

# Display the counts of each category
print(category_counts)
```
