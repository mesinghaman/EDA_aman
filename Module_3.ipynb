```python
import pandas as pd

# Function to normalize data
def normalize_rollno(data, old_min, old_max, new_min, new_max):
    """
    Normalizes the data to a given range [new_min, new_max] using the formula:

    normalized_data = ((data - old_min) / (old_max - old_min)) * (new_max - new_min) + new_min
    """
    return ((data - old_min) / (old_max - old_min)) * (new_max - new_min) + new_min


# Assuming we are normalizing the 'age' column:
old_min = df['age'].min()
old_max = df['age'].max()

# Normalizing 'age' to the range [0, 1]
df['age_normalized_0_1'] = normalize_rollno(df['age'], old_min, old_max, 0, 1)

# Normalizing 'age' to the range [-1, 1]
df['age_normalized_-1_1'] = normalize_rollno(df['age'], old_min, old_max, -1, 1)

# Normalizing 'age' to the range [-3, 3]
df['age_normalized_-3_3'] = normalize_rollno(df['age'], old_min, old_max, -3, 3)

# Displaying the DataFrame
print(df)
```
